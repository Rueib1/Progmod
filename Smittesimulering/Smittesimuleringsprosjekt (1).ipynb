##### Modell I:
En modell for smittespredning av sykdommer kan finnes ved $ I_{t+1} = I_t + aI_t $. Der I er anntal smittede individer, t er tid. Modellen viser at antall nye smittede er nå værende smittede pluss en konstant "a" ganget med antall smittede. "a" viser til hvor mange en gjennomsnittlig syk person smitter. Dette er en relativt enkelt modell for smittespredning. Den er derfor godt tillpasset til grove estimater, der man trenger en fungerende modell på kort tid. Modellen tar ikke hennsyn til: at syke folk ikke kan bli smittet igjen, at folk blir friske, at folk dør, forflyttningen til folk, vaksiner eller smittevern.

Importerer biblioteker:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

Setter opp variabler:

```python
def definisjoner (a,I,P):
    t = 0 # Tid (fungerer også som start tid)
    t_slutt = 48 # Slutt tid
    a = a # Kontaktrate
    I = I # Antall smittede
    P = P # Populasjon
    smittede = []
    tid = []
    return t, t_slutt, a, I, P, smittede, tid

t, t_slutt, a, I, P, smittede, tid = definisjoner(0.2, 100,157759)

def enkel_modell(t, t_slutt, I, a, smittede, tid):
    while t < t_slutt:
        I = I + a*I
        smittede.append(I)
        tid.append(t)
        t+=1
    return t, t_slutt, I, a, smittede, tid

start_smittede = smittede
i = I #Hvis jeg skriver "i" istedenfor "I " så fungerer fortsatt programmet
figur_nummer = 0
```

Lager en enkel modell for smittespredning ved en populasjon på 157759 individer, 100 smittede til å starte med og en kontaktrate på 0.2 per uke, det er viktig at kontakt raten har en enhet, her per uke, hvis ikke ville det ikke gitt mening, uten enhet ville smitten vært en sansynlighet for at du smittet noen, resten av livet ditt:


```python
t, t_slutt, a, I, P, smittede, tid = definisjoner(0.2, 100,157759)
t, t_slutt, I, a, smittede, tid = enkel_modell(t, t_slutt, I, a, smittede, tid)

ax = sns.scatterplot( x= tid, y = smittede)
plt.grid()
#plt.legend()
ax.set_xlabel("Tid siden start (uker)")
ax.set_ylabel("Antall smittede")
plt.title(f"Figur {figur_nummer}")
figur_nummer += 1
plt.show()
```

Figur 0 viser at utviklingen eksponensiell, vi ser at desto flere som er smittet, blir flere smittet. Det er til og med flere smittet enn det det er individer i populasjonen (flere arter har blitt smittet :p). Utforsker sammenhengen mellom antall smittede til å starte med og a.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # lager sub plots
plt.suptitle(f"Figur {figur_nummer}")


# første subplot
t, t_slutt, a, I, P, smittede, tid = definisjoner(0.2, 1000, 157759)
start_smittede = I
t, t_slutt, I, a, smittede, tid = enkel_modell(t, t_slutt, I, a, smittede, tid)
sns.scatterplot(x=tid, y=smittede, ax=axes[0, 0])
axes[0, 0].set_title(f"a = {a}, start smittede = {start_smittede}")
axes[0, 0].set_xlabel("Tid (uker)")
axes[0, 0].set_ylabel("Antall smittede")
axes[0, 0].grid()

# andre subplot
t, t_slutt, a, I, P, smittede, tid = definisjoner(0.2, 20, 157759)
start_smittede = I
t, t_slutt, I, a, smittede, tid = enkel_modell(t, t_slutt, I, a, smittede, tid)
sns.scatterplot(x=tid, y=smittede, ax=axes[0, 1])
axes[0, 1].set_title(f"a = {a}, start smittede = {start_smittede}")
axes[0, 1].set_xlabel("Tid (uker)")
axes[0, 1].set_ylabel("Antall smittede")
axes[0, 1].grid()

#tredje subplot
t, t_slutt, a, I, P, smittede, tid = definisjoner(0.5, 100, 157759)
start_smittede = I
t, t_slutt, I, a, smittede, tid = enkel_modell(t, t_slutt, I, a, smittede, tid)
sns.scatterplot(x=tid, y=smittede, ax=axes[1, 0])
axes[1, 0].set_title(f"a = {a}, start smittede = {start_smittede}")
axes[1, 0].set_xlabel("Tid (uker)")
axes[1, 0].set_ylabel("Antall smittede")
axes[1, 0].grid()

# fjerde subbplot
t, t_slutt, a, I, P, smittede, tid = definisjoner(1, 100, 157759)
start_smittede = I
t, t_slutt, I, a, smittede, tid = enkel_modell(t, t_slutt, I, a, smittede, tid)
sns.scatterplot(x=tid, y=smittede, ax=axes[1, 1])
axes[1, 1].set_title(f"a = {a}, start smittede = {start_smittede}")
axes[1, 1].set_xlabel("Tid (uker)")
axes[1, 1].set_ylabel("Antall smittede")
axes[1, 1].grid()

fig.legend(["Antall smittede"], loc="upper left")
plt.tight_layout()
plt.show()
```

Figur 1 viser at det er mange flere som blir smittet ved høyera antall smittede til å starte med, men at "brattheten" på kurven er den samme dersom a er lik. Når a varierer, ser vi at brattheten og antall smitted øker dramatisk. Dette gir mening, med tanke på at a er et tall på hvor mange en syk person gjennomsnittlig smitter per uke. Om tallet er høyt smitter personen, gjennomsnittlig, flere og det er igjen flere som kan smitte videre.
